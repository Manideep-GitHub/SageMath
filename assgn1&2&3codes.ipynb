```python
#Q1
alp = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
bet = alp[::-1]

# The Diana Cipher:
mat = [[]*26]*26
for i in range(26):
    mat[i] = bet[i:] + bet[:i]

def k(x):
    return alp.index(x)
```

#### Q10) Here power of base is (a, b) where base is (2,3) so highest power of base wrt a would be that (i, j) such that there exists no other (m, n) such that (2^i)(3^j) < (2^m)(3^n) < a

```python
# Q11 and Q12      # FOR COMPLETE SOLUTION REFER THE PDF

def gcdExtended(a, b):
    # Base Case
    if a == 0 :
        return b,0,1
             
    gcd,x1,y1 = gcdExtended(b%a, a)
     
    # Update x and y using results of recursive
    # call
    x = y1 - (b//a) * x1
    y = x1
     
    return gcd,x,y
gcdExtended(193707721, 761838257287)
```
